```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read csv file 
df = pd.read_csv("data/electricitygeneration.csv", header=None)

# shows only 5 rows
df.head() 
```

```python
df = df.replace('-', 0)
df.head()
```

```python
df.shape 
```

```python
#return the columns label of your dataframe
list(df.columns)
```

```python
df.columns = [
    'Name of Power Station',
    '2006/07',
    '2007/08',
    '2008/09',
    '2009/10',
    '2010/11',
    '2011/12',
	'2012/13',
	'2013/14',
	'2014/15'
]

df.head()
```

```python
df2 =df.drop(df.index[0])
# Convert the entire DataFrame to numbers
df2 = df2.apply(pd.to_numeric, errors='coerce')


```

```python
mean_200607 = df2["2006/07"].mean().round(2)
mean_200708 = df2["2007/08"].mean().round(2)
mean_200809 = df2["2008/09"].mean().round(2)
mean_200910 = df2["2009/10"].mean().round(2)
mean_201011 = df2["2010/11"].mean().round(2)
mean_201112 = df2["2011/12"].mean().round(2)
mean_201213= df2["2012/13"].mean().round(2)
mean_201314 = df2["2013/14"].mean().round(2)
mean_201415 = df2["2014/15"].mean().round(2)

print(f"Mean 2006/07: {mean_200607}")
print(f"Mean 2007/08: {mean_200708}")
print(f"Mean 2008/09 : {mean_200809}")
print(f"Mean 2009/10: {mean_201112}")
print(f"Mean 2010/11: {mean_201213}")
print(f"Mean 2011/12: {mean_201314}")
print(f"Mean 2012/13: {mean_201415}")
```

```python
median_200607 = df2["2006/07"].median().round(2)
median_200708 = df2["2007/08"].median().round(2)
median_200809 = df2["2008/09"].median().round(2)
median_200910 = df2["2009/10"].median().round(2)
median_201011 = df2["2010/11"].median().round(2)
median_201112 = df2["2011/12"].median().round(2)
median_201213= df2["2012/13"].median().round(2)
median_201314 = df2["2013/14"].median().round(2)
median_201415 = df2["2014/15"].median().round(2)

print(f"median 2006/07: {median_200607}")
print(f"median 2007/08: {median_200708}")
print(f"median 2008/09 : {median_200809}")
print(f"median 2009/10: {median_201112}")
print(f"median 2010/11: {median_201213}")
print(f"median 2011/12: {median_201314}")
print(f"median 2012/13: {median_201415}")
```

```python
periods = ['2006/07', '2007/08', '2008/09', '2009/10', '2010/11', '2011/12', '2012/13', '2013/14', '2014/15']
means = [mean_200607,mean_200708,mean_200809,mean_200910,mean_201011,mean_201112,median_201213,mean_201314,mean_201415]
medians = [median_200607,median_200708,median_200809,median_200910,median_201011,median_201112,median_201213,median_201314,median_201415]

x=np.arange(len(periods)) #must match length of periods

width = 0.3

plt.bar(x - width/2, means, width, color='cyan', label='Mean')
plt.bar(x + width/2, medians, width, color='orange', label='Median')

plt.xticks(x, periods, rotation=45)
plt.xlabel("Season")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.show()




```

```python
plt.plot(x, means, marker='o', label='Mean')
plt.plot(x, medians, marker='s', label='Median')
plt.xticks(rotation=45)
plt.xlabel("Season")
plt.ylabel("Energy Consumption")
plt.title("Energy Consumption: Mean vs Median by Season")
plt.legend()
plt.tight_layout()
plt.show()
```
