```python
import pandas as pd

# Load the sheet
df = pd.read_excel(
    "data/FTS_Upto_Jestha_82_83_hooiodt.xlsx",
    sheet_name="5_Imports_By_Commodity",
    skiprows=2  # skip title rows so headers line up
)

# Keep only Chapter 87 (vehicles) rows
df["HSCode"] = df["HSCode"].astype(str)
vehicles = df[df["HSCode"].str.startswith("87")].copy()

# Map the first 4 digits of the HS code to a vehicle category
category_map = {
    "8701": "Tractor",
    "8702": "Bus",
    "8703": "Car/Jeep/Van",
    "8704": "Goods Vehicle/Truck",
    "8705": "Special Purpose Vehicle",
    "8706": "Chassis",
    "8707": "Body",
    "8708": "Parts & Accessories",
    "8709": "Works Truck",
    "8711": "Motorcycle",
    "8712": "Bicycle",
    "8713": "Wheelchair/Disabled Carriage",
    "8714": "Vehicle Parts (Cycles)",
    "8715": "Baby Carriage",
    "8716": "Trailer/Other Vehicle",
}

vehicles["Category"] = vehicles["HSCode"].str[:4].map(category_map)

print(vehicles[["HSCode", "Description", "Category", "Imports_Value"]].head(10))
```
