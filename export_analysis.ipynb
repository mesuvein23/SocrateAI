```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read excel file
df = pd.read_excel("data\Current-macroeconomic-situation-Tables-Based-on-Six-Month-data-of-2082.83.xlsx", sheet_name = 17,header=None)

df.head()
```

```python
#droping 0 index columns
df = df.drop(columns =[0])
df = df.iloc[7:].reset_index(drop=True) #dropping row to 6 th 

df.head(10)
```

```python
df.describe()

print(f"Sum of 2024/25: {df['2024/25 Annual'].sum()}")     # add up the whole column → total exports that year
print(f"Mean of 2024/25: {df['2024/25 Annual'].mean()}")    # average value across commodities
print(f"Max of 2024/25: {df['2024/25 Annual'].max()}")     # the biggest single value
print(f"Min of 2024/25: {df['2024/25 Annual'].min()}") 
```
