```python
top10 = df.nlargest(10, "2024/25 Annual").sort_values("2024/25 Annual")

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.barh(top10["Commodities"], top10["2024/25 Annual"], color="steelblue")

ax.bar_label(bars, fmt="%.0f", padding=3)      # <-- the magic line

ax.set_xlabel("Imports, Annual 2024/25 (Rs. million)")
ax.set_title("Top 10 Imports From India")
plt.tight_layout()
plt.show()
```

```python
cols = ["2023/24 Annual", "2024/25 Annual",
        "2023/24 Six Months", "2024/25 Six Months", "2025/26 Six Months"]

stats = pd.DataFrame({
    "Mean": df[cols].mean(),
    "Median": df[cols].median(),
    "Std": df[cols].std(),
    "Variance": df[cols].var()
})

stats["CV"] = stats["Std"] / stats["Mean"]          # as a ratio
stats["CV %"] = stats["CV"] * 100                    # as a percentage
stats["Mean - Median"] = stats["Mean"] - stats["Median"]
stats = stats.round(2)
print(stats)
```

```python
for period, row in stats.iterrows():
    d = row["Mean - Median"]
    if d > 0:
        verdict = "right-skewed (mean > median)"
    elif d < 0:
        verdict = "left-skewed (mean < median)"
    else:
        verdict = "roughly symmetric"
    print(f"{period}: Mean - Median = {d:,.2f} → {verdict}")
```

"In every period, the mean substantially exceeds the median, confirming a right-skewed distribution: a small number of high-value commodities pull the average far above the typical export, so the median better represents the typical commodity."

```python
# calculating growth
df["growth"] = (df["2025/26 Six Months"] / df["2024/25 Six Months"] - 1) * 100
# Throw away rows with no answer
g = df.dropna(subset=["growth"]) 
g = g[(g["2025/26 Six Months"] >= 50) | (g["2024/25 Six Months"] >= 50)]   # >= Rs. 50M in either period

# Sort ascending — worst decliner at bottom, best grower at top
g = g.sort_values("growth")

# Color by sign: green for growth, red for decline
colors = ["#2e7d32" if v >= 0 else "#c62828" for v in g["growth"]]

# Plot
fig, ax = plt.subplots(figsize=(9, 8))
bars = ax.barh(g["Commodities"], g["growth"], color=colors)

ax.axvline(0, color="black", linewidth=0.8)   # zero line
ax.bar_label(bars, fmt=lambda v: f"{v:+.0f}%", padding=3, fontsize=8)
ax.set_xlabel("Six-month growth, 2024/25 → 2025/26 (%)")
ax.set_title("Exports to India: Fastest Growers and Biggest Decliners")
ax.grid(axis="x", alpha=0.3)
ax.set_xlim(g["growth"].min() * 1.25, g["growth"].max() * 1.18)  # room for labels
plt.tight_layout()
plt.show()
```

```python
periods = ["2023/24 Annual", "2023/24 Six Months", "2024/25 Annual",
              "2024/25 Six Months", "2025/26 Six Months"]
means = df[periods].mean().round(2)
medians = df[periods].median().round(2)

x = np.arange(len(periods)) #positions
width = 0.35 #bar width two bars must fit side by side

fig, ax = plt.subplots(figsize=(10, 6))
b1 = ax.bar(x - width/2, means, width, label="Mean", color="#2a78d6")
b2 = ax.bar(x + width/2, medians, width, label="Median", color="#898781")

ax.bar_label(b1, fmt=lambda v: f"{v:,.0f}", padding=3, fontsize=9)
# ax.bar_label(b2, fmt=lambda v: f"{v:,.0f}", padding=3, fontsize=9)
# median: value + percentage of mean on a second line
ax.bar_label(b2,
             labels=[f"{m:,.0f}\n({m/mn*100:.0f}% of mean)"
                     for m, mn in zip(medians, means)],
             padding=3, fontsize=8)
ax.set_xticks(x)
ax.set_xticks(x, periods)
ax.set_xticklabels(periods, rotation=20)
ax.set_ylabel("NPR million")
ax.set_title("Mean vs Median Export Value")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
six_cols = ["2023/24 Six Months", "2024/25 Six Months", "2025/26 Six Months"]

df["mean_6m"] = df[six_cols].mean(axis=1)   # across the row
df["std_6m"]  = df[six_cols].std(axis=1)
df["cv_6m"]   = df["std_6m"] / df["mean_6m"]

# Filter out tiny commodities (CV is unreliable for near-zero values)
stable = df[df["mean_6m"] >= 50].dropna(subset=["cv_6m"])

# Lowest CV = most stable
stable = stable.nsmallest(10, "cv_6m")
print(stable[["Commodities"] + six_cols + ["cv_6m"]])

plot_df = stable.sort_values("cv_6m", ascending=False)  # most stable at top after barh

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.barh(plot_df["Commodities"], plot_df["cv_6m"] * 100, color="#2e7d32")

ax.bar_label(bars, fmt="%.1f%%", padding=3)
ax.set_xlabel("Coefficient of Variation across 3 six-month periods (%)")
ax.set_title("Most Stable Export Commodities (Lowest CV)")
plt.tight_layout()
```

```python

col = "2025/26 Six Months"   # the column we suspect has an outlier (Soyabean Oil)


# 1. IQR METHOD
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Lower bound: {lower_bound:.2f}")
print(f"Upper bound: {upper_bound:.2f}")

iqr_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
print(f"\nIQR Outliers in '{col}':")
print(iqr_outliers[["Commodities", col]])


# 2. Z-SCORE METHOD
mean_val = df[col].mean()
std_val = df[col].std()

df["z_score"] = (df[col] - mean_val) / std_val

z_outliers = df[df["z_score"].abs() > 3]
print(f"\nZ-score Outliers in '{col}' (|z| > 3):")
print(z_outliers[["Commodities", col, "z_score"]])

# See the z-scores for every commodity, sorted by extremity
print("\nAll commodities ranked by |z-score|:")
print(df[["Commodities", col, "z_score"]].reindex(df["z_score"].abs().sort_values(ascending=False).index))





# 4. WHAT TO DO ABOUT IT — a version of the data WITHOUT the outlier,
#    so you can compare "with vs without" for mean/median/std
df_no_outliers = df[~df.index.isin(iqr_outliers.index)]

print(f"\nMean {col} WITH outliers:    {df[col].mean():.2f}")
print(f"Mean {col} WITHOUT outliers: {df_no_outliers[col].mean():.2f}")
print(f"Median {col} WITH outliers:    {df[col].median():.2f}")
print(f"Median {col} WITHOUT outliers: {df_no_outliers[col].median():.2f}")
```
